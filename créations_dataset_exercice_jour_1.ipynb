```python
import pandas as pd
import numpy as np
from faker import Faker

# Initialiser Faker
fake = Faker()

# Créer un DataFrame avec des doublons
np.random.seed(42)  # Pour la reproductibilité

# Nombre d'utilisateurs
n_users = 10000

# Générer des utilisateurs avec Faker
data = {
    'nom': [fake.name() for _ in range(n_users)],
    'email': [fake.email() for _ in range(n_users)],
    'date_inscription': pd.date_range(start='2020-01-01', periods=n_users, freq='H'),
    'telephone': [fake.phone_number() for _ in range(n_users)]
}

# Introduire des doublons
duplicates = np.random.choice(range(n_users), size=int(n_users * 0.1), replace=False)  # 10% de doublons
for i in duplicates:
    if i < n_users - 1:  # S'assurer de ne pas dépasser l'indice
        data['nom'][i + 1] = data['nom'][i]  # Dupliquer le nom
        data['email'][i + 1] = data['email'][i]  # Dupliquer l'email

# Créer le DataFrame
df_utilisateurs = pd.DataFrame(data)

# Afficher les premières lignes du DataFrame avant suppression des doublons
print(df_utilisateurs.head())
# Afficher le nombre de doublons
print("Nombre de doublons détectés :", df_utilisateurs.duplicated(subset=['nom', 'email']).sum())

# En option, sauvegarder le DataFrame sous forme de fichier CSV
df_utilisateurs.to_csv('exercice1.csv', index=False)

```

```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialiser Faker
fake = Faker()

# Nombre d'utilisateurs
n_users = 10000

# Fonction pour générer des dates dans des formats différents et y ajouter des chaînes supplémentaires
def generate_random_date():
    # Choisir un format aléatoire
    format_choice = random.choice(['DD/MM/YYYY', 'MM-DD-YYYY', 'YYYY/MM/DD', 'YYYY-MM-DD'])
    if format_choice == 'DD/MM/YYYY':
        date = fake.date_of_birth(minimum_age=18, maximum_age=70).strftime('%d/%m/%Y')
    elif format_choice == 'MM-DD-YYYY':
        date = fake.date_of_birth(minimum_age=18, maximum_age=70).strftime('%m-%d-%Y')
    elif format_choice == 'YYYY/MM/DD':
        date = fake.date_of_birth(minimum_age=18, maximum_age=70).strftime('%Y/%m/%d')
    else:
        date = fake.date_of_birth(minimum_age=18, maximum_age=70).strftime('%Y-%m-%d')

    # Ajouter aléatoirement des chaînes supplémentaires (par exemple "08h20M00S" ou "UTC")
    if random.random() < 0.3:  # 30% de chances d'ajouter une chaîne
        # Générer des heures, minutes, secondes aléatoires ou indicateurs supplémentaires
        time_part = random.choice([
            f"{random.randint(0, 23):02d}h{random.randint(0, 59):02d}M{random.randint(0, 59):02d}S",  # Ex : 08h20M00S
            f"{random.randint(0, 23):02d}h{random.randint(0, 59):02d}M",  # Ex : 14h35M
            f"{random.randint(0, 23):02d}:{random.randint(0, 59):02d}:{random.randint(0, 59):02d}",  # Ex : 14:35:10
        ])
        timezone_part = random.choice([' UTC', ' GMT', ' CET', ' PST', ' EST'])  # Différents indicateurs de fuseau horaire

        # Ajouter soit l'heure, soit le fuseau, ou les deux
        extra_info = random.choice([time_part, timezone_part, time_part + timezone_part])
        date += " " + extra_info

    return date

# Générer des utilisateurs avec Faker
data = {
    'nom': [fake.name() for _ in range(n_users)],
    'email': [fake.unique.email() for _ in range(n_users)],  # Assurer l'unicité des emails
    'age': [fake.random_int(min=18, max=70) for _ in range(n_users)],  # Générer les âges
    'date_inscription': [generate_random_date() for _ in range(n_users)],  # Dates incohérentes
    'telephone': [fake.phone_number() for _ in range(n_users)]
}

# Créer le DataFrame
df_utilisateurs = pd.DataFrame(data)

# Afficher les premières lignes du DataFrame
print(df_utilisateurs.head())

# En option, sauvegarder le DataFrame sous forme de fichier CSV
df_utilisateurs.to_csv('exercice 3.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialiser Faker
fake = Faker()

# Nombre de transactions
n_transactions = 1000

# Fonction pour générer des dates dans des formats variés
def generate_random_date():
    # Choisir un format aléatoire
    format_choice = random.choice(['DD/MM/YYYY', 'MM-DD-YYYY', 'YYYY/MM/DD', 'YYYY-MM-DD'])
    if format_choice == 'DD/MM/YYYY':
        date = fake.date_between(start_date='-5y', end_date='today').strftime('%d/%m/%Y')
    elif format_choice == 'MM-DD-YYYY':
        date = fake.date_between(start_date='-5y', end_date='today').strftime('%m-%d-%Y')
    elif format_choice == 'YYYY/MM/DD':
        date = fake.date_between(start_date='-5y', end_date='today').strftime('%Y/%m/%d')
    else:
        date = fake.date_between(start_date='-5y', end_date='today').strftime('%Y-%m-%d')

    # Ajouter des chaînes supplémentaires (ex: heures, fuseaux horaires)
    if random.random() < 0.3:  # 30% de chances d'ajouter une chaîne
        time_part = random.choice([
            f"{random.randint(0, 23):02d}h{random.randint(0, 59):02d}M{random.randint(0, 59):02d}S",  # Ex : 08h20M00S
            f"{random.randint(0, 23):02d}h{random.randint(0, 59):02d}M",  # Ex : 14h35M
            f"{random.randint(0, 23):02d}:{random.randint(0, 59):02d}:{random.randint(0, 59):02d}",  # Ex : 14:35:10
        ])
        timezone_part = random.choice([' UTC', ' GMT', ' CET', ' PST', ' EST'])  # Différents fuseaux horaires
        date += " " + random.choice([time_part, timezone_part, time_part + timezone_part])

    return date

# Générer des clients avec Faker
data = {
    'client_id': [fake.uuid4() for _ in range(n_transactions)],  # UUID pour chaque client
    'montant': [round(random.uniform(1, 5000), 2) for _ in range(n_transactions)],  # Montants aléatoires entre 1 et 5000
    'date_transaction': [generate_random_date() for _ in range(n_transactions)],  # Dates incohérentes
    'mode_paiement': [random.choice(['Carte', 'carte', 'CARTE', 'Virement', None]) for _ in range(n_transactions)]  # Modes de paiement variés
}

# Créer le DataFrame
df_transactions = pd.DataFrame(data)

# Ajouter quelques valeurs aberrantes dans la colonne 'montant'
for _ in range(10):  # Ajouter 10 valeurs aberrantes
    index = random.randint(0, n_transactions - 1)
    df_transactions.loc[index, 'montant'] = random.uniform(10000, 20000)  # Valeurs très élevées

# Afficher les premières lignes du DataFrame
print(df_transactions.head())

# En option, sauvegarder le DataFrame sous forme de fichier CSV
df_transactions.to_csv('OPT 2.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialiser Faker
fake = Faker()

# Nombre de commandes
n_commandes = 1000

# Liste de statuts avec variations
statuts_variations = ['Livrée', 'livrée', 'LIVREE', 'En cours', 'annulée', 'non livrée', None]

# Fonction pour générer des dates dans des formats variés
def generate_random_date():
    # Choisir une date aléatoire dans le passé
    return fake.date_between(start_date='-1y', end_date='today')

# Fonction pour générer une date de livraison
def generate_random_livraison_date():
    if random.random() < 0.3:  # 30% de chances de générer une date de livraison dans le futur
        return fake.date_between(start_date='today', end_date='+30d')
    else:
        return generate_random_date()

# Générer des données
data = {
    'commande_id': [fake.uuid4() for _ in range(n_commandes)],
    'client_id': [fake.uuid4() for _ in range(n_commandes)],
    'montant': [round(random.uniform(10, 1000), 2) for _ in range(n_commandes)],
    'date_commande': [generate_random_date() for _ in range(n_commandes)],
    'status': [random.choice(statuts_variations) for _ in range(n_commandes)],
    'livraison_date': [generate_random_livraison_date() for _ in range(n_commandes)],
}

# Créer le DataFrame
df_commandes = pd.DataFrame(data)

# Ajouter des doublons aléatoires
for _ in range(50):  # Ajouter 50 doublons
    df_commandes = pd.concat([df_commandes, df_commandes.sample(n=1)])

# Ajouter des valeurs manquantes
for col in ['montant', 'status', 'livraison_date']:
    df_commandes.loc[df_commandes.sample(frac=0.1).index, col] = None  # 10% de valeurs manquantes

# Ajouter quelques valeurs aberrantes dans la colonne 'montant'
for _ in range(10):  # Ajouter 10 valeurs aberrantes
    index = random.randint(0, df_commandes.shape[0] - 1)
    df_commandes.loc[index, 'montant'] = random.uniform(5000, 20000)  # Valeurs très élevées

# Afficher les premières lignes du DataFrame
print(df_commandes.head())

# En option, sauvegarder le DataFrame sous forme de fichier CSV
df_commandes.to_csv('OPT 3.csv', index=False)
```
