Exercice 1 : Graphique en ligne des ventes mensuelles

Consigne : Visualisez l'évolution des ventes mensuelles d'un produit sur un an.

Exercice 10 : Scatter plot avec annotations

Consigne : Créez un scatter plot pour montrer la relation entre la taille et le poids de 100 individus, puis ajoutez une annotation pour la valeur maximale de poids.

```python
# Données fictives
taille = np.random.normal(170, 10, 100)
poids = taille * 0.5 + np.random.normal(0, 5, 100)

```

```python
# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(taille, poids, color='orange')
plt.title("Relation entre taille et poids")
plt.xlabel("Taille (cm)")
plt.ylabel("Poids (kg)")

# Annotation de la valeur maximale de poids
indice_max = np.argmax(poids)
plt.annotate('Poids max', xy=(taille[indice_max], poids[indice_max]),
             xytext=(taille[indice_max]+2, poids[indice_max]+2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

Exercice 14 : Subplots avec plusieurs courbes

Consigne : Utilisez des sous-graphiques pour afficher l'évolution des ventes de 3 produits différents sur une période de 12 mois.

```python
# Données fictives
mois = np.arange(1, 13)
ventes_produit_a = np.random.randint(100, 200, 12)
ventes_produit_b = np.random.randint(150, 250, 12)
ventes_produit_c = np.random.randint(80, 180, 12)

```

```python
# Subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].plot(mois, ventes_produit_a, color='blue', marker='o')
axs[0].set_title("Ventes Produit A")
axs[0].set_xlabel("Mois")
axs[0].set_ylabel("Ventes")

axs[1].plot(mois, ventes_produit_b, color='green', marker='o')
axs[1].set_title("Ventes Produit B")
axs[1].set_xlabel("Mois")
axs[1].set_ylabel("Ventes")

axs[2].plot(mois, ventes_produit_c, color='red', marker='o')
axs[2].set_title("Ventes Produit C")
axs[2].set_xlabel("Mois")
axs[2].set_ylabel("Ventes")

plt.tight_layout()
plt.show()
```

Exercice 15 : Graphique en bandes empilées (Stacked Area Plot)

Consigne : Visualisez les ventes cumulées de plusieurs produits au cours de chaque trimestre.

```python
# Données fictives
trimestres = np.arange(1, 9)
produit_a = np.random.randint(50, 100, size=8)
produit_b = np.random.randint(30, 80, size=8)
produit_c = np.random.randint(20, 70, size=8)
```

```python
# Stacked Area Plot
plt.figure(figsize=(10, 6))
plt.stackplot(trimestres, produit_a, produit_b, produit_c, labels=['Produit A', 'Produit B', 'Produit C'], colors=['skyblue', 'orange', 'green'])
plt.title("Ventes cumulées par produit")
plt.xlabel("Trimestre")
plt.ylabel("Ventes")
plt.legend(loc='upper left')
plt.show()
```

Exercice 16 : Annotations et mise en forme avancée

Consigne : Visualisez l'évolution des ventes mensuelles sur un an et ajoutez des annotations pour les valeurs maximales et minimales.

```python
# Données fictives
mois = np.arange(1, 13)
ventes = np.random.randint(80, 160, 12)


```

```python
# Graphique avec annotations
plt.figure(figsize=(10, 6))
plt.plot(mois, ventes, marker='o', color='dodgerblue')
plt.title("Ventes mensuelles sur un an")
plt.xlabel("Mois")
plt.ylabel("Ventes")
plt.grid(True)

# Annotations
max_vente = np.max(ventes)
min_vente = np.min(ventes)
mois_max = mois[np.argmax(ventes)]
mois_min = mois[np.argmin(ventes)]

plt.annotate('Max', xy=(mois_max, max_vente), xytext=(mois_max+0.2, max_vente+5),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)
plt.annotate('Min', xy=(mois_min, min_vente), xytext=(mois_min+0.2, min_vente-10),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)
plt.show()
```
