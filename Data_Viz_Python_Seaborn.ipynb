```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Exercice 1 : Histogramme simple avec Seaborn

Consigne : Utilisez Seaborn pour créer un histogramme montrant la distribution des âges dans un groupe de 200 personnes.

```python
# Données fictives
ages = np.random.randint(18, 65, 200)

```

```python

# Histogramme
plt.figure(figsize=(8, 5))
sns.histplot(ages, bins=10, color='skyblue')
plt.title("Distribution des âges")
plt.xlabel("Âge")
plt.ylabel("Fréquence")
plt.show()
```

Exercice 2 : Bar Plot pour la répartition des ventes

Consigne : Visualisez les ventes par région dans un bar plot avec Seaborn.

```python
# Données fictives
regions = ['Nord', 'Sud', 'Est', 'Ouest']
ventes = [120, 80, 100, 90]
data = pd.DataFrame({'Région': regions, 'Ventes': ventes})
```

Exercice 4 : Scatter Plot avec Seaborn

Consigne : Créez un scatter plot pour montrer la relation entre la taille et le poids de 100 individus.

```python
# Données fictives
taille = np.random.normal(170, 10, 100)
poids = taille * 0.5 + np.random.normal(0, 5, 100)
data = pd.DataFrame({'Taille': taille, 'Poids': poids})

```

```python

# Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Taille', y='Poids', color='orange')
plt.title("Relation entre taille et poids")
plt.xlabel("Taille (cm)")
plt.ylabel("Poids (kg)")
plt.show()
```

Exercice 5 : Bar Plot avec une catégorisation

Consigne : Utilisez un bar plot pour visualiser la moyenne des pourboires (tip) par jour de la semaine dans le dataset tips, en séparant les données par genre (sex).

```python
# Chargement des données
tips = sns.load_dataset("tips")
```

```python

# Bar Plot avec catégorisation
plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x='day', y='tip', hue='sex', palette="pastel")
plt.title("Pourboire moyen par jour et par genre")
plt.xlabel("Jour")
plt.ylabel("Pourboire moyen")
plt.show()
```

Exercice 7 : Heatmap pour afficher une matrice de corrélation

Consigne : Utilisez une heatmap pour montrer la corrélation entre les différentes variables du dataset tips.

```python
# Sélection des colonnes numériques pour la matrice de corrélation
numerical_tips = tips.select_dtypes(include=['float64', 'int64'])
correlation = numerical_tips.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Matrice de corrélation des variables")
plt.show()
```

```python
# Chargement des données
tips = sns.load_dataset("tips")

# Transformation des colonnes non numériques en variables numériques avec one-hot encoding
tips_encoded = pd.get_dummies(tips, drop_first=True)

# Calcul de la matrice de corrélation
correlation = tips_encoded.corr()

# Heatmap de la matrice de corrélation
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Matrice de corrélation avec variables numériques et catégorielles encodées")
plt.show()
```

Exercice 8 : KDE Plot pour la densité

Consigne : Créez un KDE plot pour visualiser la distribution de la variable total_bill dans le dataset tips.

```python
# KDE Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(data=tips, x="total_bill", fill=True, color="purple", alpha=0.5)
plt.title("Densité des factures totales")
plt.xlabel("Montant de la facture")
plt.show()
```

Exercice 10 : Swarm Plot avec jitter

Consigne : Utilisez un swarm plot pour visualiser la distribution des pourboires (tip) par jour avec un effet de jitter.

```python
# Swarm Plot
plt.figure(figsize=(8, 5))
sns.swarmplot(data=tips, x="day", y="tip", color="blue", alpha=0.6)
plt.title("Pourboires par jour avec effet de dispersion (jitter)")
plt.xlabel("Jour")
plt.ylabel("Pourboire")
plt.show()
```

Exercice 11 : Facet Grid pour visualiser des sous-ensembles de données

Consigne : Utilisez un Facet Grid pour afficher la distribution des pourboires (tip) par jour en fonction du temps (time : déjeuner ou dîner).

```python
# Facet Grid
g = sns.FacetGrid(tips, col="day", hue="time", height=4, aspect=0.9)
g.map(sns.histplot, "tip", kde=True)
g.add_legend()
g.set_axis_labels("Pourboire", "Fréquence")
g.fig.suptitle("Distribution des pourboires par jour et par période", y=1.05)
plt.show()
```

Exercice 12 : Line Plot avec intervalle de confiance

Consigne : Utilisez un line plot pour afficher l'évolution du pourcentage de pourboires (tip/total_bill) en fonction de la taille du groupe (size) dans le dataset tips, avec un intervalle de confiance.:

```python
# Ajout d'une colonne pourcentage de pourboire
tips['tip_percentage'] = tips['tip'] / tips['total_bill'] * 100

# Line Plot avec intervalle de confiance
plt.figure(figsize=(10, 6))
sns.lineplot(data=tips, x="size", y="tip_percentage", ci="sd", marker='o', color='blue')
plt.title("Pourcentage de pourboire en fonction de la taille du groupe")
plt.xlabel("Taille du groupe")
plt.ylabel("Pourcentage de pourboire (%)")
plt.grid(True)
plt.show()
```

Exercice 13 : Joint Plot avec la distribution et la régression

Consigne : Utilisez un joint plot pour afficher la relation entre total_bill et tip, en montrant également les distributions marginales et une ligne de régression.

```python
# Joint Plot avec régression
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg", height=8, marginal_kws=dict(bins=30, fill=True))
plt.suptitle("Relation entre le montant total et le pourboire", y=1.05)
plt.show()
```

Exercice 14 : Catplot pour des visualisations catégorielles avancées

Consigne : Utilisez un catplot pour afficher les pourboires (tip) par jour et par période (time), en séparant par smoker (fumeur ou non-fumeur).

```python
# Catplot
sns.catplot(data=tips, x="day", y="tip", hue="smoker", col="time", kind="bar", height=5, aspect=0.8, palette="Set2")
plt.suptitle("Pourboires par jour, période et habitude de fumer", y=1.05)
plt.show()
```
